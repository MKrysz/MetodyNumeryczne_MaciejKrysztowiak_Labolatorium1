***Zadanie 4.***  
Częstą praktyką szczególnie w dziedzinie analizy danych, statystyce, uczeniu maszynowym, itp. jest tworzenie na podstawie biblioteki [Matplotlib](https://matplotlib.org/) własnych szablonów wykresów stworzonych na podstawie funkcji. Najprostszy przykład to przeładowanie funkcji plot tak by wykres miał czerwony kolor:

```python
def my_plot(x,y,xlabel,ylabel,title,label):
    if x.shape != y.shape or  min(x.shape)==0:
        return None
    fig, ax = plt.subplots()
    ax.plot(x, y, 'r', label=label)
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    ax.legend()
    return fig

x = np.linspace(-10,10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
my_plot(x,y,'x','y','Funkcja sin', 'sin(x)')
```

**Zadanie 5**  
Za pomocą funkcji *compare\_plot* rozwiąż graficznie równanie: $f(x) = g(x)$, gdzie:  
$f(x) = x+2$  
$g(x) = x^2 -2sin(x) + 3$

```python
import main
def f(x):
    return x+2
def g(x):
    return x**2 - 2* np.sin(x) + 3

x = np.linspace(-10, 10)
y1 = f(x)
y2 = g(x)

main.compare_plot(x, y1, x, y2, 'x', 'y', 'untilted', 'x+2', '$x^2 -2sin(x) + 3$')
```

Z wykresu powyżej widać, że funkcje przecinają się dwa razy w okolicach (0, 2.5), korzystając z funkcji xlim i ylim przybliżam wykres

```python
main.compare_plot(x, y1, x, y2, 'x', 'y', 'untilted', 'x+2', '$x^2 -2sin(x) + 3$')
plt.xlim(0, 2.5)
plt.ylim(0, 5)
plt.grid(True)
```

Następnie przybliżam pierwszy punkt przecięcia

```python
main.compare_plot(x, y1, x, y2, 'x', 'y', 'untilted', 'x+2', '$x^2 -2sin(x) + 3$')
plt.xlim(0.3, 0.5)
plt.ylim(2.2, 2.5)
plt.grid(True)
```

Pierwszy punkt przecięcia znajduje się w okolicach (0.42, 2.43)

```python
main.compare_plot(x, y1, x, y2, 'x', 'y', 'untilted', 'x+2', '$x^2 -2sin(x) + 3$')
plt.xlim(1.5, 1.7)
plt.ylim(3.5, 3.7)
plt.grid(True)
```

***Zadanie 7.***  
Za pomocą funkcji *parallel\_plot*  i przedstaw na jednym z nich [Spirale logarytmiczną](https://pl.wikipedia.org/wiki/Spirala_logarytmiczna) w szerokim przedziale, a w drugim w okolicy zera.

```python
theta1 = np.linspace(-10*np.pi, 10*np.pi, int(1e5))
theta2 = np.linspace(-1*np.pi, 1*np.pi, int(1e5))
a = 1.5
b = 0.2
x1 = a * np.exp(b*theta1) * np.cos(theta1)
y1 = a * np.exp(b*theta1) * np.sin(theta1)
x2 = a * np.exp(b*theta2) * np.cos(theta2)
y2 = a * np.exp(b*theta2) * np.sin(theta2)
main.parallel_plot(x1, y1, x2, y2, 'X', 'Y', 'x', 'y', 'logarythmic spiral', '|')
```
